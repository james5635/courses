```python
import numpy as np
import matplotlib.pyplot as plt
# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Create a figure
fig = plt.figure()

# Create a subplot
ax = fig.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')

# Create two subplots and unpack the output array immediately
ax1, ax2 = fig.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)

# Create four polar Axes and access them through the returned array
axes = fig.subplots(2, 2, subplot_kw=dict(projection='polar'))
axes[0, 0].plot(x, y)
axes[1, 1].scatter(x, y)

# Share an X-axis with each column of subplots
fig.subplots(2, 2, sharex='col')

# Share a Y-axis with each row of subplots
fig.subplots(2, 2, sharey='row')

# Share both X- and Y-axes with all subplots
fig.subplots(2, 2, sharex='all', sharey='all')

# Note that this is the same as
fig.subplots(2, 2, sharex=True, sharey=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function y = x^1000
def f(x):
    return x**1000

# Generate x values close to 1 to avoid overly large y-values
x = np.linspace(0.8, 1.2, 400)
y = f(x)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$y = x^{1000}$', color='purple')
plt.axvline(1, color='gray', linestyle='--', label='x = 1')
plt.axvline(2, color='red', linestyle='--', label='x = 2')
plt.yscale('log')  # log scale to handle large values
plt.title('Graph of $y = x^{1000}$ (log scale)')
plt.xlabel('x')
plt.ylabel('y (log scale)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range for x
# Choose a small range around 1 to avoid excessively large numbers
# For example, from 0.9 to 1.1
x = np.linspace(0.9, 1.1, 500)

# Calculate y values
# Be aware that for x > 1, y will grow very rapidly
y = x**1000

# Create the plot
plt.figure(figsize=(10, 6))

# Plot on a logarithmic y-scale
# plt.semilogy(x, y)
# plt.ylim(1e10,1e200)

plt.plot(x, y, 'r')
# Add labels and title
plt.xlabel('x')
plt.ylabel('y (log scale)')
plt.title('Plot of y = x^1000 (log scale)')

# Add a grid for better readability
plt.grid(True, which="both", ls="-", color='0.7')
plt.tight_layout()
# Show the plot
plt.show()
```

```python
# Re-import after code execution environment reset
import numpy as np
import matplotlib.pyplot as plt

# Example data
x = np.linspace(1, 5, 5)
y = 10 ** x  # y = [10, 100, 1000, 10000, 100000]

# Create two subplots side by side
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Left: Regular linear scale
axs[0].plot(x, y, marker='o')
axs[0].set_title('Linear Scale')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].grid(True)

# Right: Logarithmic y-scale
axs[1].plot(x, y, marker='o')
axs[1].set_yscale('log')  # Only changes display, not data
axs[1].set_title('Logarithmic y-scale')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y (log scale)')
axs[1].grid(True, which='both')

plt.tight_layout()
plt.show()
```
