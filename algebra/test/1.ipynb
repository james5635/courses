```python
import matplotlib.pyplot as plt
import numpy as np
xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 2 * (xmax - xmin)
x = np.linspace(xmin,xmax, 50)
print(x)
fig,ax = plt.subplots()
plt.axis([xmin, xmax, ymin,ymax])
plt.plot([xmin,xmax], [0,0], 'b')
plt.plot([0,0], [ymin,ymax], 'b')

y = 2*x + 1
print(y)
plt.plot(x,y, 'r-')
# plt.plot([1], [5], 'bo' )
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xmin, xmax = -10, 10
points = 100
x = np.linspace(xmin, xmax, points)
y = np.sin(x)

# plt.figure(figsize=(6, 4))
plt.plot(x, y, 'r-', label=f'{points} points')
plt.title("Sine Curve with Few Points")
# plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure()

plt.subplot(2, 2, 1)
plt.plot([1, 2,3])
plt.title("Plot 1")

plt.subplot(2, 2, 2)
plt.plot([3, 2, 1])
plt.title("Plot 2")

plt.subplot(2, 2, 3)
plt.plot([1, 1, 1])
plt.title("Plot 3")

plt.subplot(2, 2, 4)
plt.plot([0, 0, 0])
plt.title("Plot 4")

plt.tight_layout()  # prevent overlapping
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.random.uniform(0, 2 * np.pi, 1024)
# data2d = np.sin(t)[:, np.newaxis] * np.cos(t)[np.newaxis, :]
data2d = np.sin(t)[:, np.newaxis] * np.exp(t)[np.newaxis, :]
print(t)
print(data2d)
fig, ax = plt.subplots()
im = ax.imshow(data2d)
ax.set_title('Pan on the colorbar to shift the color mapping\n'
             'Zoom on the colorbar to scale the color mapping')

fig.colorbar(im, ax=ax, label='Interactive colorbar')

plt.show()
```
