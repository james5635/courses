# using numpy library instead of a loop

```python
import matplotlib.pyplot as plt
import numpy as np

xmin = -10
xmax = 10
ymin = -10
ymax = 10

points = 10 * (xmax - xmin)
x= np.linspace(xmin, xmax, points)
print(len(x))
fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax])
plt.plot([xmin,xmax], [0,0], 'b')
plt.plot([0,0], [ymin,ymax], 'b')

y1 = 2*x
plt.plot(x, y1)

y2 = x**2 -3
plt.plot(x,y2)

plt.show()
```
