```python
import matplotlib.pyplot as plt
x1 = 0
y1 = 55
x2 = 6
y2 = 76

m = (y2-y1)/(x2-x1)
b = y1 - m*x1
print("y = ", m, "x + ", b)

xmin = -0
xmax = 10
ymin = 0
ymax = 150

y3 = m*xmin + b
y4 = m*xmax + b 

fig,ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax])
plt.plot([xmin,xmax], [0,0], 'b')
plt.plot([0,0], [ymin,ymax], 'b')

ax.set_xlabel ( "time")
ax.set_ylabel ("population")
ax.grid(True)

plt.plot([xmin,xmax], [y3,y4], 'r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x1 = 5
y1 = 10
x2 = 25
y2 = 4

m = (y2-y1)/(x2-x1)
b = y1 - m*x1
print("y = ", m, "x + ", b)

xmin = 0
xmax = 50
ymin = 0
ymax = 15

y3 = m*xmin + b
y4 = m*xmax + b 

fig,ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax])
plt.plot([xmin,xmax], [0,0], 'b')
plt.plot([0,0], [ymin,ymax], 'b')


ax.set_xlabel ( "time")
ax.set_ylabel ("profit")
ax.grid(True)
ax.set_xticks(np.arange(xmin,xmax, 2))
ax.set_yticks(np.arange(ymin,ymax, 1))

plt.plot([xmin,xmax], [y3,y4], 'r')
plt.show()
```
