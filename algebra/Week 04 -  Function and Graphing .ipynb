# Display axis lines

```python
import matplotlib.pyplot as plt
xmin = -10
xmax = 10
ymin = -10
ymax = 10
fig,ax = plt.subplots()
plt.axis([xmin, xmax, ymin,ymax])
plt.plot([xmin,xmax], [0,0], 'b')
plt.plot([0,0], [ymin,ymax], 'b')
plt.show()
```

# Plot one point

```python
import matplotlib.pyplot as plt
xmin = -10
xmax = 10
ymin = -10
ymax = 10
fig,ax = plt.subplots()
plt.axis([xmin, xmax, ymin,ymax])
plt.plot([xmin,xmax], [0,0], 'b')
plt.plot([0,0], [ymin,ymax], 'b')
plt.plot([5], [4], 'or')
plt.show()
```

# Plot serveral points as a function

```python
import matplotlib.pyplot as plt
xmin = -10
xmax = 10
ymin = -10
ymax = 10
fig,ax = plt.subplots()
plt.axis([xmin, xmax, ymin,ymax])
plt.plot([xmin,xmax], [0,0], 'b')
plt.plot([0,0], [ymin,ymax], 'b')

for x in range(10):
    y = 0.5*x + 1
    plt.plot([x], [y], 'or')

plt.show()
```

# graph and table of (x,y) values

```python
import matplotlib.pyplot as plt
xmin = -10
xmax = 10
ymin = -10
ymax = 10
fig,ax = plt.subplots()
plt.axis([xmin, xmax, ymin,ymax])
plt.plot([xmin,xmax], [0,0], 'b')
plt.plot([0,0], [ymin,ymax], 'b')

print('x \t y')
for x in range(-5,10):
    y = 0.5*x + 1
    plt.plot([x], [y], 'or')
    print(x,"\t",y)

plt.show()
```

```python
import matplotlib.pyplot as plt
xmin = -10
xmax = 10
ymin = -10
ymax = 10
fig,ax = plt.subplots()
plt.axis([xmin, xmax, ymin,ymax])
plt.plot([xmin,xmax], [0,0], 'b')
plt.plot([0,0], [ymin,ymax], 'b')

print('x \t y')
for x in range(xmin,xmax):
    y = 0.5*x + 1
    plt.plot([x], [y], 'or')
    print(x,"\t",y)

plt.show()
```
