A good use of natural log: how long will it take for an investment to double in value?
Start with the continuous growth formula:
$ A = Pe^{rt} $   
if the amount is double, then annuity is two times the principle:
$ 2P = Pe^{rt} $  
Divide both sides by P and get this:
$ 2 = e^{rt} $  
Because of the base e, take the natural log of both sides and get this:
$ ln(2) = rt $  
Then divide by r to solve for t:
$ \frac{ln(2)}{r} = t $

```python
import math
r = .02
t = round(math.log(2)/r)
print("It will double in ", t, "years")
```

Graphing

```python
import matplotlib.pyplot as plt
import numpy as np
import math

xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 10 * (xmax-xmin)
x = np.linspace(xmin,xmax, points)
x2 = np.linspace(.01, xmax, points)

fig,ax = plt.subplots()
plt.axis([xmin, xmax, ymin, ymax])
plt.plot([xmin,xmax], [0,0], 'b')
plt.plot([0,0], [ymin, ymax], 'b')

y1 = math.e**x
plt.plot(x, y1)

y2 = np.log(x2)
plt.plot(x2, y2)

y3 = x
plt.plot(x, y3)
```

Convert to scientific notation

```python
import math
a = .0005
exp = math.floor(math.log(a, 10))
n = round(a/(10**exp), 2)

if n>= 10:
    exp = exp + 1
    n = a/(10**exp)
print(a, " = ", n, "* 10**", exp)
```
