# Loop through function input

```python
import matplotlib.pyplot as plt
xmin = -10
xmax = 10
ymin = -10
ymax = 10
fig,ax = plt.subplots()
plt.axis([xmin, xmax, ymin,ymax])
plt.plot([xmin,xmax], [0,0], 'b')
plt.plot([0,0], [ymin,ymax], 'b')

for x in range(10):
    y = 0.5*x + 1
    plt.plot([x], [y], 'or')

plt.show()
```

# Use a array as input

```python
import matplotlib.pyplot as plt
import numpy as np
xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 2 * (xmax - xmin)
x = np.linspace(xmin,xmax, points)
print(x)
fig,ax = plt.subplots()
plt.axis([xmin, xmax, ymin,ymax])
plt.plot([xmin,xmax], [0,0], 'b')
plt.plot([0,0], [ymin,ymax], 'b')

y = 2*x + 1
print(y)
plt.plot(x,y, 'r')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 2 * (xmax - xmin)
x = np.linspace(xmin,xmax, points)
fig,ax = plt.subplots()
plt.axis([xmin, xmax, ymin,ymax])
plt.plot([xmin,xmax], [0,0], 'b')
plt.plot([0,0], [ymin,ymax], 'b')
ax.set_xlabel("x values")
ax.set_ylabel("y values")
ax.set_title("Some Graph")
ax.grid(True)
ax.set_xticks(np.arange(xmin,xmax, 1))
# print(np.arange(xmin,xmax, 1))
ax.set_yticks(np.arange(ymin,ymax, 1))
# print(np.arange(ymin,ymax, 2))
y = 2*x + 1
plt.plot(x,y, label='y=2x+1')
plt.plot([4],[6], 'ro', label='point')
plt.plot(x, 3*x, label="steeper line")
plt.legend()
plt.show()
```
