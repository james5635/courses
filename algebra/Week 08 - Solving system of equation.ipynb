# Simplest way to solve a system of equations. set each equation  equal to zero

```python
from sympy import *

x, y = symbols("x y")
first = 2 * x + y - 1
second = x - 2 * y + 7
print(linsolve([first, second], (x, y)))
```

```python
from sympy import *
x,y = symbols('x y')

first = 2*x + y -1
second = x - 2*y + 7

solution = linsolve([first, second], (x,y))
x_solution = solution.args[0][0]
y_solution = solution.args[0][1]
# print(solution)
# print(solution.args)
print("(", x_solution, ",", y_solution, ")")
```

# Factor to change from "0 = " to "y = " preparing to graph

```python
from sympy import *
from sympy.plotting import plot

var('x y')
first = 2*x + y - 1
second = x - 2*y + 7

solution = linsolve([first, second], [x,y])
x_solution = solution.args[0][0]
y_solution = solution.args[0][1]
print("Solution = (", x_solution, ",", y_solution, ")")

y_first = Eq(first, 0)

y1 = solve(y_first, y)

y_second = Eq(second, 0)
y2 = solve(y_second, y)

print("y = ", y1[0])
print("y = ", y2[0])

x = symbols("x")
plot(y1[0], y2[0], (x, -10,10))
```
