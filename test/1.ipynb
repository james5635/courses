```python
import matplotlib.pyplot as plt
import numpy as np

xmin, xmax = -10, 10
points = 100
x = np.linspace(xmin, xmax, points)
y = np.sin(x)

# plt.figure(figsize=(6, 4))
plt.plot(x, y, 'r-', label=f'{points} points')
plt.title("Sine Curve with Few Points")
# plt.grid(True)
plt.legend()
plt.show()
```
